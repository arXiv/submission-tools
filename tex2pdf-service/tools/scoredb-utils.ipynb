```python
import json
import sqlite3

import pandas as pd
```

```python
from sqlite3 import Connection
```

```python
def score_db(score_path: str) -> Connection:
    """Open scorecard database."""
    db = sqlite3.connect(score_path)
    db.execute("create table if not exists score (source varchar primary key, outcome TEXT, pdf varchar, status int, success bool)")
    db.execute("create table if not exists touched (filename varchar primary key)")
    return db
```

```python
sdb = score_db("score.db")
```

```python
sdbtai = score_db("TAI-DB/score.db")
```

```python
used_clsfiles = {}
for index, row in dfx.iterrows():
    # print(row['source'])
    filejson = row['clsfiles']
    # strip /usr/local/texlive/202[34] prefix
    files = json.loads(filejson)
    files = list(set(files))
    for f in files:
        # print(f)
        if f in used_clsfiles:
            used_clsfiles[f] += 1
        else:
            used_clsfiles[f] = 1
```

```python
dict(sorted(used_clsfiles.items(), key=lambda x: x[1], reverse=True))
```

```python
used_styfiles = {}
for index, row in dfx.iterrows():
    # print(row['source'])
    filejson = row['styfiles']
    # strip /usr/local/texlive/202[34] prefix
    files = json.loads(filejson)
    files = list(set(files))
    for f in files:
        # print(f)
        if f in used_styfiles:
            used_styfiles[f] += 1
        else:
            used_styfiles[f] = 1
```

```python
arxiv_used_files = {}
only_fancyhdr_used = 0
for index, row in dfx.iterrows():
    # print(row['source'])
    filejson = row['arxivfiles']
    # strip /usr/local/texlive/202[34] prefix
    files = [f.removeprefix("/usr/local/texlive/2023/").removeprefix("/usr/local/texlive/2024/") for f in json.loads(filejson)]
    files = list(set(files))
    if len(files) == 1 and files[0] == "texmf-arxiv/tex/latex/fancyhdr/fancyhdr.sty":
        only_fancyhdr_used += 1
    for f in files:
        # print(f)
        if f in arxiv_used_files:
            arxiv_used_files[f] += 1
        else:
            arxiv_used_files[f] = 1
```
